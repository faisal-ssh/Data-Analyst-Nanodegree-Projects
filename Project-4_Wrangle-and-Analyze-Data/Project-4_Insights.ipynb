# WeRateDogs Twitter Account Insights
>WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog.<br>
The dataset was collected from Twitter API using Tweepy library, and then analyzed to come up with the following insights.

## Insight 1:
- It appears from the scatter plot and heatmap that there is a strong positive linear relatioship between the public metrics (retweet, reply, like, and quote counts)<br> with values over 0.7 which means when one increases so do the others. As for the rating, it has no linear relationship at all to the public metrics<br>and this is expected becasue of the nature of the funny rating system of the account that rates dogs without standards.


## Insight 2:
- It looks like from both plots that almost all tweets come from one source 'Twitter for iPhone'.<br> Also among dog stages, 'pupper' seems to have the highest number of tweets while 'floofer' is the lowest.<br> In addition, many tweets didn't specify the dog stage in their text indicated by the second plot.

## Insight 3:
- We can see that for the period from 2015-11 until 2016-05 (half a year), nearly all tweets (with dog stage specifed)<br> were about dogs in pupper stage. Then afterwards, other tweets of different dog stages started to show up.<br> Also, the chart shows that doggo stage is generally prefered having high avergae number of likes and retweets compared to others.

## Insight 4:
- Clearly from the word cloud, dog stage pupper is the most frequent word and it confirms to what we showed in the bar plot<br> that most tweets are about pupper dog stage. Other words like: pet, dog, meet, hello are ones of the most occurances as well.<br> This is expected as this account usually greet the dog which is a pet and then gives it a rating.
